```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

img = cv2.imread("messi.jpg")
cv2_imshow(img)

#for grayimage
gray_image = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#for inverting the image
invt_img = cv2.bitwise_not(gray_image)

#for smoothing of image
img_smt = cv2.GaussianBlur(invt_img,(21,21),sigmaX=0,sigmaY=0)

def dodgeV2(x, y):
  return cv2.divide(x, 255-y, scale=256)

final_img=dodgeV2(gray_image,img_smt)
cv2_imshow(final_img)
```
